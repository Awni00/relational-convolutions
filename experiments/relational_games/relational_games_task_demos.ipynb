# Demos of 'Relational Games' Tasks

```python
import numpy as np
import matplotlib.pyplot as plt
import os; os.environ['TF_CPP_MIN_LOG_LEVEL'] = '2'
import data_utils
```

```python
data_path = '../../data/relational_games'
```

```python
def load_task_sample(filename, sample_size):
    with np.load(filename) as data:
        imgs = data['images']
        labels = data['labels']

    obj_seqs = data_utils.get_obj_seq(imgs)
    labels = np.squeeze(labels)

    sample = np.random.choice(len(imgs), sample_size, replace=False)

    imgs_sample, obj_seqs_sample, labels_sample = imgs[sample], obj_seqs[sample], labels[sample]
    del imgs, obj_seqs, labels, data
    return imgs_sample, obj_seqs_sample, labels_sample
```

## 'Between'

```python
filename = f'{data_path}/1task_between_stripes.npz'

imgs, obj_seqs, labels = load_task_sample(filename, sample_size=32)
```

```python
sample = np.random.choice(np.where(labels==1)[0])

fig, axs = plt.subplots(nrows=3, ncols=3, figsize=(5, 5))
for j, ax in enumerate(axs.flatten()):
    ax.imshow(obj_seqs[sample][j])
    ax.axis('off')
    ax.set_title(f'object {j}')

print(f'label: {bool(labels[sample])}')
```

```python
sample = np.random.choice(np.where(labels==0)[0])

fig, axs = plt.subplots(nrows=3, ncols=3, figsize=(5, 5))
for j, ax in enumerate(axs.flatten()):
    ax.imshow(obj_seqs[sample][j])
    ax.axis('off')
    ax.set_title(f'object {j}')

print(f'label: {bool(labels[sample])}')
```
